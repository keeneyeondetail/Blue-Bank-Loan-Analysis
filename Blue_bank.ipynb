```python
import json
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
json_file=open("loan_data_json.json")
```

```python
data=json.load(json_file)
data
```

```python
dataset=pd.DataFrame(data)
dataset
```

```python
dataset["purpose"].unique()
```

```python
dataset.describe()
```

```python
dataset["income"]=np.exp(dataset["log.annual.inc"]) 
```

```python
length=len(dataset)
ficocat=[]
for x in range(0,length):
    category=dataset["fico"][x]
    if category >=300 and category <400:
        cat="very-poor"
    elif category >=400 and category <600:
        cat="poor"
    elif category >=601 and category <660:
        cat="fair" 
    elif category >=660 and category <700:
        cat="good"   
    elif category>700:
        cat="excellent"
    else:
        cat="unknown"
    ficocat.append(cat) 
    
    
    
```

```python
ficosat=pd.Series(ficocat)
```

```python
dataset["fico_ctegory"]=ficosat
dataset
```

```python
dataset.loc[dataset["int.rate"]>.12,"int_type"]="High"
```

```python
dataset.loc[dataset["int.rate"]<=.12,"int_type"]="Low"
```

```python
dataset.to_csv("Blue_bank.csv")
```
